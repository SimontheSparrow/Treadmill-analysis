Sprawdźmy jakie modele są najpopularniejsze wśród mężczyzn i kobiet.


```python
pd.crosstab(df['Product'], df['Gender'])
```

```python
sns.countplot(x='Product', hue='Gender', data=df);
```

Modele TM195  i TM498 są równo rozłożone pośród kobiet i mężczyzn. Model TM798 jest wyraźnie bardziej preferowany przez mężczyzn.
Sprawdźmy jakie grupy wiekowe korzystają z bieżni.

```python
sns.boxenplot(x=df['Gender'], y=df['Age']);
```

Dodatkowo możemy sprawdzić stan cywilny oraz dochody badanych.

```python
pd.crosstab(df['Product'], df['MaritalStatus'])
```

```python
df.hist(by='Gender', column = 'Income');
```

```python
sns.displot(df['Age'], kde=True);
```

```python
sns.pairplot(df);
```

Zobaczmy jakie są korelacje między danymi.

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, annot=True);
```

```python
sns.scatterplot(x='Age', y='Income', hue='Product', data=df);
```

Osoby zarabiające najwięcej preferują model TM798, pozostałe dwa modele są rozłożone pośród badanych z niższymi dochodami. Wiek nie ma znaczenia.
